```python
from keras.datasets import mnist
import numpy as np

(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
print(y_train.shape)
```

### TOP 

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
for j in range(len(x_train)):
    for i in range(14,28):
        x_train[j][i] = 0
        
for j in range(len(x_test)):
    for i in range(14,28):
        x_test[j][i] = 0
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for i in range(3):
    plt.imshow(x_train[i],cmap = 'gray')
    plt.show()
    
for i in range(3):
    plt.imshow(x_test[i],cmap = 'gray')
    plt.show()
```

```python
print(x_train.shape)
print(x_test.shape)
np.save('top_x_train',x_train)
np.save('top_x_test',x_test)
```
